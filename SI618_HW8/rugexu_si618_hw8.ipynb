### Q6. For 30 points: Let's try to more concretly analyse if the odds of a legendary pokemon winning was related to the generation they were introduced in. Load the combat encounter dataset and join it with the pokemon dataset to identify the generation and legendary status of each pokemon in an encounter. Considering encounters between legendary pokemon introduced in a given generation and regular pokemon of any generation, create a cross tabulation of generation against the wins/losses of legendary pokemon.
* Are there * statistically significant* differences in the odds of a legendary pokemon winning based on the generation it was introduced in? ($\alpha=0.05$). What do you observe?
* Does this remain true if we limit the encounters to pokemon introduced in the same generation?

```python
pokemon_info = pokemon[["#", "Generation", "Legendary"]]

combats['#'] = combats['First_pokemon']

combats1 = combats.merge(pokemon_info, how='left', on='#')
combats1 = combats1.rename(columns={"Generation":"First_Generation", "Legendary":"First_Legendary"})

combats1['#'] = combats1['Second_pokemon']

combats2 = combats1.merge(pokemon_info, how='left', on='#')
combats2 = combats2.rename(columns={"Generation":"Second_Generation", "Legendary":"Second_Legendary"})

combats2['#'] = combats2['Winner']

combats_final = combats2.merge(pokemon_info, how='left', on='#')
combats_final = combats_final.rename(columns={"Generation":"Winner_Generation", "Legendary":"Winner_Legendary"})

combats_final = combats_final[(((combats_final['First_Legendary'] == True)
                                & (combats_final['Second_Legendary'] == False)) |
                              ((combats_final['First_Legendary'] == False) & (combats_final['Second_Legendary'] == True)))]
combats_final
```

```python
ct = pd.crosstab(combats_final.Winner_Legendary, combats_final.Winner_Generation)
ct
```

```python
chi2, p, dof, ex = chi2_contingency(ct)
p
```

It can be seen that the p-value obtained by chi-square test is very small, less than 0.05. This result indicates that we can reject the null hypothesis that there is no statistical significance in the odds of a legendary Pokemon winning in all generations. Therefore, there is a relationship between whether the Pokemon is legendary and whether it would win.

```python
combats_final2 = combats_final[(combats_final['First_Generation'] == combats_final['Second_Generation'])]
ct = pd.crosstab(combats_final2.Winner_Legendary, combats_final2.Winner_Generation)
ct
```

```python
chi2, p, dof, ex = chi2_contingency(ct)
p
```

This time, the p-value obtained by chi-square test is not so small, larger than 0.05. Thus we can not reject the null hypothesis that there is no statistical significance in the odds of a legendary Pokemon winning in the same generations. We cannot say whether there is a relationship between whether the Pokemon is legendary and whether it would win in the same generation.
