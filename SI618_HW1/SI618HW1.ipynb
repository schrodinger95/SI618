## Length

Q4: You will want to create a new column with the integer value of the number of pages. (If you remove rows in this process, please state why.)

I remove the rows whose book_pages is NaN. Then the column can be changed into integer directly.

```python
book_df2 = book_df.dropna(subset = ['book_pages'])
book_df2['book_pages_count'] = book_df2['book_pages'].str.replace(' pages', '').astype('int')
book_df2
```

Q5: What is the median number of pages?

```python
book_df2['book_pages_count'].median()
```

Q6: What are the minimum and maximum numbers of pages?

```python
book_df2['book_pages_count'].min()
```

```python
book_df2['book_pages_count'].max()
```

Q10: Discretize (i.e. round down) the ratings. The resulting ratings should have one of the following values: 1,2,3,4 or 5.

```python
book_df['book_rating_rounded'] = book_df['book_rating'].astype('int')
book_df
```

Q11: For each of the discretized ratings (1,2,3,4,5), what is the average number 
of authors?

```python
book_df['book_authors_count'] = book_df['book_authors'].str.count('\|') + 1
book_df.groupby(['book_rating_rounded'])['book_authors_count'].mean()
```

Q12: How about the average number of reviews?

## Genres

Q13: Create a new DataFrame, exploding the rows with multiple genres such that it is one row per genre/book.

```python
book_df3=book_df.drop('genres', axis=1).join(book_df['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('genres'))
book_df3
```

Q14: What is the average rating of books from different genres? What is the median?

```python
book_df3.groupby(['genres'])['book_rating'].mean()
```

```python
book_df3.groupby(['genres'])['book_rating'].median()
```

Q15: What is the average number of authors of books from different genres? What is the median?

```python
book_df3.groupby(['genres'])['book_authors_count'].mean()
```

```python
book_df3.groupby(['genres'])['book_authors_count'].median()
```
